```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
pca = PCA()
pca_values = pca.fit_transform(wine_normal)
```

```python
pca_values
```

```python
pca = PCA(n_components = 14)
pca_values = pca.fit_transform(wine_normal)
```

```python
pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set()


plt.plot(var*100) # scree plot
plt.show()
```

```python
plt.plot(var1,color="red")
```

# Using Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(finalDf, method='complete'))
```

```python
hc = AgglomerativeClustering(n_clusters = 3, linkage= 'complete')
```

```python
y_hc = hc.fit_predict(finalDf)
clusters_data = pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
clusters_data
```

```python
clusters_data['Clusters'].value_counts()
```

# Using K Means Clustering

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

by pca

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
klist = []

for i in range(1,11):
    kmeans = KMeans( n_clusters = i )
    kmeans.fit(finalDf)
    klist.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),klist)
plt.title('Elbow Plot')
plt.show()
```

as we can see elbow plot suggests us 3 would be the best value for clustering

```python
model = KMeans(n_clusters=3)
model.fit(finalDf)
```

```python
model.labels_
```

```python
md = pd.Series(model.labels_)
finalDf['clust'] = md
```

```python
finalDf['clust'].value_counts()
```
